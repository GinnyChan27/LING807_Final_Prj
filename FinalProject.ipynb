### Text 1

```python
with open ("TestFile1-5c29947f1e67d78e27b6d330.txt", "r", encoding='utf-8') as f:
    text1 = f.read()
```

```python
doc1 = nlp(text1)
```

```python
doc1._.coref_chains.print()
```

```python
# check spacy morphology/ POS tagging property 
for token in doc1:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.morph)
```

```python
# print a token to see which chain it belongs to/ who it is refering to
print(doc1._.coref_chains.resolve(doc1[1700]))
```

```python
#### print out token number for easier reference
for i, token in enumerate(doc1):
    print(f"Token {i}: {token.text}")
```

```python
from spacy import displacy
```

```python
displacy.render(doc1, style="dep", options= {"compact": True}, jupyter=True)
```

```python
options = {"ents": ["PERSON"], 
          "colors": {"PERSON": "lightsteelblue"}}

displacy.render(doc1, style="ent", options=options, jupyter=True)
```

#### Text 1 brief analysis notes

- SpaCy recognise an obvious plural over singular + singular antecedents. 
E.g Chain 0: "After weeks of online flirting, Patrick Bardos was en route to meet Anne Marie Cerato for their first date at a coffee shop in downtown Toronto."
Weeks > Patrick + Anne for 'their'. Even though based on relevance, Patrick+Anne should more likely be the antecedents that 'weeks'.
- Error even in simple co-reference, can be exceptions or marginal errors due to data cleaning errors. Only a real error if it is a constant mistake (maybe pointing towards that spaCy does not have gendered labelling?)
E.g Chain 4: 'woman' is considered in 'Bardos' Chain which spaCy has been consistently connecting/labelling as male pronouns.
- Some vague nouns were not considered an entity and thus not included in the chain. (usually those using 'who' afterwards)
E.g 'guests' is not included in any chain. "...attend to his peeved party guests, who spent the night listening to him rave about this woman he just met. "
- Pronouns are wrongly assigned to animals (Why does spaCy make this connection? Does it have pre-labelled gender for animals rather than referring to them as 'it/they')
E.g Chain 28: 'cat' is assigned male pronouns even though those were referring to another entity in the text.
- spaCy has difficulty in identifying and wrongly connecting abstract obj/concepts with pronouns
E.g Chain 31, 35, 42
